# Resize the entire "With_background" database

```python
# New size (selected so that all the characters are completely in) 
new_width, new_height =  700, 800

path = "Data/With_background/"
image_prob = []

for image in tqdm(os.listdir(path)):
    # Do this code or store image name of there is a problem
    try:
        image_path = path + image
         Opens image in RGB
        im = Image.open(image_path)  
        # Image size
        width, height = im.size
    
        # pixels index of new image 
        left = (width - new_width)/2
        top = (height - new_height)/2
        right = (width + new_width)/2
        bottom = (height + new_height)/2

        # Crop the image
        im = im.crop((left, top, right, bottom))

        # saves image with its news dimensions
        im.save(image_path)

        image_prob.append(image)
                
    except:
        image_prob.append(image) 
```
